Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.

ANS_2:-

A scatter plot is a type of plot or mathematical diagram using Cartesian coordinates to display values for typically two variables for a set of data. If the points are color-coded, one additional variable can be displayed. The data are displayed as a collection of points, each having the value of one variable determining the position on the horizontal axis and the value of the other variable determining the position on the vertical axis.

```python
#CODE:-

import numpy as np
import matplotlib.pyplot as plt


np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.figure()
plt.scatter(x, y)


plt.title('Scatter Plot of Generated Data')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

ANS_3:-

The subplot() function in Matplotlib is used to create multiple plots in a single figure. It allows you to arrange multiple plots in a grid, enabling you to compare different datasets side by side or display various aspects of the same dataset. Each subplot can be accessed individually and customized as needed.

```python
#CODE:-

import numpy as np
import matplotlib.pyplot as plt

x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

plt.figure(figsize=(10, 8))

plt.subplot(2, 2, 1)  
plt.plot(x1, y1, 'r')  
plt.title('Line Plot 1')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')


plt.subplot(2, 2, 2)
plt.plot(x2, y2, 'g') 
plt.title('Line Plot 2')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')


plt.subplot(2, 2, 3)
plt.plot(x3, y3, 'b')  
plt.title('Line Plot 3')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.subplot(2, 2, 4)
plt.plot(x4, y4, 'm')  
plt.title('Line Plot 4')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

plt.tight_layout()
plt.show()

```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

ANS_4:-

A bar plot (or bar chart) is a graphical display of data using bars of different heights or lengths. Each bar represents a category of data, and the height or length of the bar is proportional to the values they represent. Bar plots can be plotted vertically or horizontally.

->Bar plots are used to compare different categories of data. The height or length of the bars makes it easy to see differences and trends at a glance.

->They provide a clear and straightforward way to present categorical data, making it easier for the audience to understand.


```python
#CODE:-

import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

plt.figure(figsize=(10, 5))
plt.bar(company, profit, color='blue')
plt.title('Company Profit (Vertical Bar Plot)')
plt.xlabel('Company')
plt.ylabel('Profit')
plt.show()

plt.figure(figsize=(10, 5))
plt.barh(company, profit, color='green')
plt.title('Company Profit (Horizontal Bar Plot)')
plt.xlabel('Profit')
plt.ylabel('Company')
plt.show()

```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

ANS_5:-

A box plot (or box-and-whisker plot) is a graphical representation that summarizes the distribution of a dataset. It displays the dataset's minimum, first quartile (25th percentile), median (50th percentile), third quartile (75th percentile), and maximum. Box plots also show potential outliers in the data.

-> Box plots provide a concise summary of the distribution of a dataset, including its central tendency and spread.

-> They are useful for comparing distributions between different groups or datasets.

-> They help visualize the spread of data and potential skewness, which can be useful for understanding data behavior.

```python
#CODE:-

import numpy as np
import matplotlib.pyplot as plt

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

plt.figure(figsize=(8, 6))
plt.boxplot([box1, box2], labels=['Box 1', 'Box 2'])
plt.title('Box Plot of Two Datasets')
plt.xlabel('Dataset')
plt.ylabel('Values')
plt.show()

```
