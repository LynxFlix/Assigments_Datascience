Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

ANS_1:_

Seaborn is a powerful Python visualization library based on Matplotlib that provides a high-level interface for drawing attractive and informative statistical graphics. Here are five types of plots you can create using Seaborn, along with their uses:

1. Scatter Plot (sns.scatterplot):-

    -> Use: Displays the relationship between two continuous variables. It's useful for identifying correlations, trends, and patterns in data.

2. Bar Plot (sns.barplot):-

    -> Use: Compares the average value of a continuous variable across different categories. It's useful for visualizing comparisons and aggregations.
    
3. Histogram (sns.histplot):-

    ->Use: Displays the distribution of a single continuous variable by dividing it into bins. It's useful for understanding the frequency distribution and spread of the data.
    
4. Box Plot (sns.boxplot):-

    -> Use: Shows the distribution of a continuous variable across different categories by displaying quartiles and potential outliers. It's useful for comparing distributions and spotting outliers.
    
5. Heatmap (sns.heatmap):-

    -> Use: Visualizes the magnitude of values in a matrix-like format using color gradients. It's useful for displaying correlation matrices and patterns in data.
    
    


Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
"timepoint" and y = "signal" for different events and regions.

ANS_2:-

```python
#CODE:-

import seaborn as sns
import matplotlib.pyplot as plt

fmri = sns.load_dataset('fmri')

plt.figure(figsize=(12, 8))
sns.lineplot(data=fmri, x='timepoint', y='signal', hue='event', style='region', markers=True)

plt.title('Signal Over Time for Different Events and Regions')
plt.xlabel('Timepoint')
plt.ylabel('Signal')

plt.show()

```

Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
'pclass', y = 'age' and y = 'fare'.
Note: pclass, age, and fare are columns in the titanic dataset.

ANS_3:-

```python
#CODE:-

import seaborn as sns
import matplotlib.pyplot as plt

titanic = sns.load_dataset('titanic')

plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
sns.boxplot(data=titanic, x='pclass', y='age')
plt.title('Box Plot of Age by Pclass')
plt.xlabel('Pclass')
plt.ylabel('Age')

plt.subplot(1, 2, 2)
sns.boxplot(data=titanic, x='pclass', y='fare')
plt.title('Box Plot of Fare by Pclass')
plt.xlabel('Pclass')
plt.ylabel('Fare')

plt.tight_layout()

plt.show()

```

Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.

ANS_4:-

```python
#CODE:-

import seaborn as sns
import matplotlib.pyplot as plt

diamonds = sns.load_dataset('diamonds')

plt.figure(figsize=(12, 8))
sns.histplot(data=diamonds, x='price', hue='cut', multiple='stack', bins=30)

plt.title('Histogram of Diamond Prices by Cut')
plt.xlabel('Price')
plt.ylabel('Frequency')

plt.show()
```

Que 6: Use the "flights" dataset from seaborn to plot a heatmap.

ANS_6:-

```python
#CODE:-

import seaborn as sns
import matplotlib.pyplot as plt

flights = sns.load_dataset('flights')

flights_pivot = flights.pivot('month', 'year', 'passengers')

plt.figure(figsize=(12, 8))
sns.heatmap(data=flights_pivot, cmap='YlGnBu', annot=True, fmt='d')

plt.title('Number of Passengers per Month and Year')
plt.xlabel('Year')
plt.ylabel('Month')

plt.show()
```
